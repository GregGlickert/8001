```python
from scipy import signal
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#making the first order system
first_num = [1]
first_den = [2,1]
first_order_sys = signal.TransferFunction(first_num, first_den)
#step reponse code
y,t = signal.step(first_order_sys)
plt.plot(y,t)
plt.xlabel('Time')
plt.ylabel('response')
plt.title('step response')
plt.show()
#impluse reponse
y,t = signal.impulse(first_order_sys)
plt.plot(y,t)
plt.xlabel('Time')
plt.ylabel('response')
plt.title('impulse response')
plt.show()
#sinusoidal
t = np.linspace(0,50,100)
u = np.sin(t)
t,yout,xout = signal.lsim(first_order_sys,u,t)
plt.plot(t,yout)
plt.xlabel('Time')
plt.ylabel('response')
plt.title('sin response')
plt.show()
```

```python
#making the second order system
second_num = [1]
second_den = [4,2,1]
second_order_sys = signal.TransferFunction(second_num, second_den)
#step reponse code
y,t = signal.step(second_order_sys)
plt.plot(y,t)
plt.xlabel('Time')
plt.ylabel('response')
plt.title('step response')
plt.show()
#impluse reponse
y,t = signal.impulse(second_order_sys)
plt.plot(y,t)
plt.xlabel('Time')
plt.ylabel('response')
plt.title('impulse response')
plt.show()
#sinusoidal
t = np.linspace(0,50,100)
u = np.sin(t)
t,yout,xout = signal.lsim(second_order_sys,u,t)
plt.plot(t,yout)
plt.xlabel('Time')
plt.ylabel('response')
plt.title('sin response')
plt.show()
```

```python
# periodogram stuff

fs = nfft = 512
t = np.arange(1000) / fs

# added 3 sin waves with a freq of 50, 100, and 150 together
x = np.sin(2*np.pi*50*t) + np.sin(2*np.pi*100*t) + np.sin(2*np.pi*150*t)

freq, psd = signal.periodogram(x,fs,nfft=nfft)

#ploting
plt.plot(t,x)
plt.ylabel('response')
plt.xlabel('time')
plt.title('3 sin waves added together')
plt.show()
plt.plot(freq, psd)
plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')
plt.title('periodogram')
plt.show()
```
